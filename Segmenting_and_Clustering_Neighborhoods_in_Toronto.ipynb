<h1>Get Data From Wiki</h1>

```python
targetsource = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
data=BeautifulSoup(targetsource,'lxml')
print(data.title)

from IPython.display import display_html
tab = str(data.table)
display_html(tab,raw=True)
```

<h1>Data to DataFrame </h1>

```python
data = pd.read_html(tab)
dataframe=data[0]
dataframe.head()
```

<h1>Data Prep</h1>

```python
prepdata = dataframe[dataframe.Borough != 'Not assigned']

secondprepdata = prepdata.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
secondprepdata.reset_index(inplace=True)

secondprepdata['Neighborhood'] = np.where(secondprepdata['Neighborhood'] == 'Not assigned',secondprepdata['Borough'], secondprepdata['Neighborhood'])

secondprepdata
```

<h1>Copy Lat/Log Data to Data </h1>

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

<h1>Combine Data</h1>

```python
FinalData = pd.merge(secondprepdata, lat_lon, on='Postal Code')
FinalData.head()
```

<h1>Explore and cluster the neighborhoods in Toronto. You can decide to work with only boroughs that contain the word Toronto and then replicate the same analysis we did to the New York City data</h1>

```python
exploredata = FinalData[FinalData['Borough'].str.contains('Toronto',regex=False)]
exploredata
```

<h1>Visualize the neighborhoods and clustering</h1>

```python
map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(exploredata['Latitude'],exploredata['Longitude'],exploredata['Borough'],exploredata['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='red',
    fill=True,
    fill_color='#212122',
    fill_opacity=0.4,
    parse_html=False).add_to(map)
map
```
